```python
# BOX COX TRANSFORMATION FOR SALES AMOUNT 
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

# Load the data from the Excel file
file_path = r"C:\Users\HP\Untitled Folder 1\data_thu.xlsx"  # Replace with your file path
data = pd.read_excel(file_path)

# Filter out non-positive values from the 'Sales Amount' column
positive_data = data[data['Sales Amount'] > 0]

# Ensure there's enough data left for transformation
if not positive_data.empty:
    # Perform the Box-Cox transformation
    transformed_data, best_lambda = stats.boxcox(positive_data['Sales Amount'])

    # Plotting
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))
    ax[0].hist(positive_data['Sales Amount'], bins=30)
    ax[0].set_title("Original Data of Sales Amount")
    ax[1].hist(transformed_data, bins=30)
    ax[1].set_title("Box-Cox Transformed - Sales Amount")
    plt.show()
else:
    print("No positive data left after filtering. Can't apply Box-Cox transformation.")
```
